```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,2*np.pi,2*np.pi/1000) #create x=[0,...,2*pi] 
y = 5.5 * np.cos(2*x) + 5.5  # Equation a) 
z = 0.02 * np.exp(x)   # Equation b) 
w = (0.25 * x * x) + (0.1 * np.sin(10 * x)) # Equation c) 
plt.xlim([0,2*np.pi]) 
plt.ylim([-1,10])
plt.plot(x,y, label="y = 5.5 * np.cos(2*x) + 5.5")  # Plot a)
plt.plot(x,z, label="z = 0.02 * np.exp(x)")  # Plot b)
plt.plot(x,w, label="w = (0.25 * x * x) + (0.1 * np.sin(10 * x))")  # Plot c)
plt.xlabel('Time in ASTR 119')  #label x axis
plt.ylabel('Measures of Awesomeness')  #label y axis
plt.legend()
plt.show()
```
